## Data Analyst vs Data Scientist


### Findings from analyzing 2000 Indeed Job Postings in 10 U.S. cities

_New York City, Boston, Austin, Chicago, Philadelphia, San Diego, San Francisco, Seattle, Washington DC, Portland_

### Files

1. **scraper.ipynb** scrapes 2000 jobs from Indeed with 'data analyst' and 'data scientist' using BeautifulSoup and Selenium
2. **text_processing.py** processes text from the job description: tokenize the text, remove stopwords, check term frequency
3. **plot_data.py** plotting functions, make word clouds, plot skills by categories
4. **data_cleaning.ipynb** cleans scraped data and puts them in dataframes with their frequencies, then sae them to csv
5. **indeed_findings.ipynb** presents findings

### Data Analyst jobs are more about Business Analytics than Machine Learning. 


Other top keywords include: <i> information, product, reporting, leadership, financial, maintain, solution, client, database, communication skill, model,..<i>


```python
###Data analyst####
titles = ['data analyst Boston', 'data analyst New York city', 'data analyst Seattle',
        'data analyst Portland', 'data analyst Austin', 'data analyst Chicago',
         'data analyst Philadelphia', 'data analyst San Diego', 'data analyst Washington DC',
         'data analyst San Francisco']
posting_da =  string_wc(titles)
make_wc(posting_da)
```

### Data Scientist jobs are more about Machine Learning than Business or Analytics

```python
##### data scientist analysis ####
ds_titles = ['data scientist New York city', 'data scientist Boston','data scientist Austin', 
          'data scientist Chicago', 'data scientist Philadelphia',
          'data scientist San Diego', 'data scientist San Francisco', 
          'data scientist Seattle', 'data scientist Washington DC','data scientist Portland']
posting_ds = string_wc(ds_titles)
make_wc(posting_ds)
```
